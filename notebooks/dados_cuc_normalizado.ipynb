```python
import os
import pandas as pd
import numpy as np

from sklearn import preprocessing
from sklearn.model_selection import train_test_split

import json
import joblib
```

```python
#12x12

path = "C:/Users/Everton/Documents/Mestrado/projeto-final-tensorflow/dados"
myFiles = os.listdir(path)

#dados completos contendo todos os dicionarios criados (total 108)
dados = pd.DataFrame(columns = 
                     ['Bocal', 
                      'Pressão', 
                      'Quebra Jato',
                      'Velocidade Vento',
                      'Direção Vento', 
                      'Distância',
                      'CUC'])

# ler os dados de cada arquivo na pasta
for file in myFiles:
    data = pd.read_excel(path + "/" + file, header=None, sheet_name=None)
    #dicionario com os dados descritores (um para cada aba de cada arquivo)
    descritores = {}
    #lendo cada uma das abas e colocando os descritores nos dados
    for key in data.keys():
        descritores["Bocal"] = float(data[key][1][19][0:3].replace(",","."))
        descritores["Pressão"] = float(data[key][1][20][0:3].replace(",","."))
        descritores["Quebra Jato"] = float(data[key][1][21][0:3].replace(",","."))
        
        descritores["Velocidade Vento"] = str(int(data[key][1][23]*100000)/100000.0)
        descritores["Direção Vento"] = str(int(data[key][1][25]*100000)/100000.0)
        descritores["Distância"] = "12x12"
                                                      
#         if data[key][1][17].hour < 10:
#             descritores["horario_experimento"] = "INICIO_MANHA"
#         elif data[key][1][17].hour >= 10 and data[key][1][17].hour < 13:
#             descritores["horario_experimento"] = "FIM_MANHA"
#         elif data[key][1][17].hour >= 13 and data[key][1][17].hour < 16:
#             descritores["horario_experimento"] = "INICIO_TARDE"
#         else:
#             descritores["horario_experimento"] = "FIM_TARDE"
        
#         descritores["mes_experimento"] = str(data[key][1][22].month)
        
#         descritores["vazao_inicial"] = data[key][3][20]
#         descritores["vazao_final"] = data[key][3][21]
#         descritores["vazao_total"] = data[key][3][22]
        
        #calculo cuc
        df = pd.DataFrame(np.zeros((16, 16)))
        matrix = data[key][0:16][0:16]
        
        k = 4
        w = 4
        for i in range(0,12):
            if w > 14:
                w = 4
            for j in range(0,12):
                df.at[i,j] = 0 + matrix.at[k,w]
                w = w + 1
            k = k+1

        k = 4
        w = 0
        for i in range(0,12):
            if w > 11:
                w = 0
            for j in range(4,16):
                df.at[i,j] = df.at[i,j] + matrix.at[k,w]
                w = w + 1
            k = k + 1

        k = 0
        w = 4
        for i in range(4,16):
            if w > 14:
                w = 4
            for j in range(0,12):
                df.at[i,j] = df.at[i,j] + matrix.at[k,w]
                w = w + 1
            k = k + 1

        k = 0
        w = 0
        for i in range(4,16):
            if w > 11:
                w = 0
            for j in range(4,16):
                df.at[i,j] = df.at[i,j] + matrix.at[k,w]
                w = w + 1
            k = k + 1
            
        soma = 0.0
        media = df.values.mean()
        
        for i in range(4, 12):
            for j in range(4,12):
                valor = df.at[i,j]
                valor_abs = abs(valor-media)
                final = valor_abs/(256*media)
                soma = soma+final
        
        cuc = (1.0-soma)*100.0
        
        descritores["CUC"] = cuc
        dados = dados.append(descritores, ignore_index = True)
```
